```python
import pandas as pd
df = pd.read_csv('data/fulldata.csv')
```

```python
df.head(2)
```

```python
clickbait_set = df[ df["label"] == 1 ]
non_clickbait_set = df[df["label"] == 0 ]
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize=(10, 5), dpi=100)
clickbait_views_log = clickbait_set["number_of_buttons"].apply(np.log).replace([-np.inf], 0)
non_clickbait_views_log = non_clickbait_set["number_of_buttons"].apply(np.log).replace([-np.inf], 0)
plt.hist(
    [clickbait_views_log, non_clickbait_views_log], 
    bins=10, 
    color=["red", "blue"], 
    label=["Clickbait", "Non-clickbait"], 
    alpha=0.5
)
plt.title("Tags distribution")
plt.xlabel("log")
plt.ylabel("samples")
plt.legend(loc='upper right')
plt.savefig("views-distribution-log", bbox_inches='tight')
plt.show()
```
